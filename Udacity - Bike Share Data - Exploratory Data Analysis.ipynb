# Question 1: 
What is the longest bike-share Trip duration in New York, Chicago, and Washington? 

```r
# Histogram displaying the Trip Duration for New York bike share

ggplot(aes(x = Trip.Duration), data = subset(ny, !is.na(Trip.Duration))) + 
    geom_histogram(binwidth = 70, color = '#63666A', fill = '#FF9E1B') +
    scale_x_continuous(limits = c(0,950), breaks = seq(0, 1000, 50)) 
```

```r
# Histogram displaying the Trip Duration for Chicago bike share

ggplot(aes(x = Trip.Duration), data = subset(chi, !is.na(Trip.Duration))) + 
    geom_histogram(binwidth = 70, color = '#63666A', fill = '#FF9E1B') +
    scale_x_continuous(limits = c(0,950), breaks = seq(0, 1000, 50)) 
```

```r
# Histogram displaying the Trip Duration for Washington bike share

ggplot(aes(x = Trip.Duration), data = subset(wash, !is.na(Trip.Duration))) + 
    geom_histogram(binwidth = 70, color = '#63666A', fill = '#FF9E1B') +
    scale_x_continuous(limits = c(0,950), breaks = seq(0, 1000, 50)) 
```

Analyzing the trip duration for New york, we can see that trip durations that lasted between 300 and 350 minutes had the highest count with a count of roughly 2500. This much higher than any other bike share trip duration in New York.

Analyzing the trip duration for Chicago, we can see that trip durations that lasted between 300 and 350 minutes had the highest count with a count of roughly 2200. This much higher than any other bike share trip duration in Chicago.

Analyzing the trip duration for Washington, we can see that trip durations the lasted between 300 and 350 minutes had the highest count with a count of roughly 2000. This much higher than any other bike share trip duration in Washington.

Based on this the longest trip duration in New York, Chicago, and Washington are between 300 and 350 minutes.

# Question 2:
Which Gender used bike-share more often?


```r
# Bar Chart for New York bike share

ggplot(aes(x = Gender), data = subset(ny, !is.na(Gender))) + 
    geom_bar(binwidth = 70, color = '#63666A', fill = '#FF9E1B') 
```

```r
# Statistics for New York bike share

by(ny$Trip.Duration, ny$Gender, summary)
```

```r
# Bar Chart for Chicago bike share

ggplot(aes(x = Gender), data = subset(chi, !is.na(Gender))) + 
    geom_bar(binwidth = 70, color = '#63666A', fill = '#FF9E1B') 
```

```r
# Statistics for Chicago bike share

by(chi$Trip.Duration, ny$Gender, summary)
```

If we look at the bar chart for New York it looks like that males use bike share more often than females. However,  compare the mean for both genders for female the mean is 886.7 and for male the mean is 773.5. Also, the 3rd QU for male is lower (916.0) compared to female 3rd Qu (1086.6). Based on this information it looks like the fameles use Bike Share more often than males in New York.

If we look at the bar chart for Chicago it looks like that males use bike share more often than females. if  compare the mean for both genders, for female the mean is 928.8 and for male the mean is 936.9. However, the 3rd QU for male is lower (1128.0) compared to female 3rd Qu (1124.0). Based on this information it looks like the fameles use Bike Share more often than males in Chicago.

If we look at the bar chart for Washington it looks like that washington has no Gender related data. But based on the results from the New York and Chicago Females will use bike share more often than males.

Overall in all three cities Femals use bike share more often than Males.

# Question 3

Which Birth Year had the highest subscription to use Bike-Share?

```r
ggplot(aes(x = 'user.type'), data = subset(ny, !is.na(Birth.Year))) +
    geom_bar(binwidth = 70, color = '#63666A', fill = '#FF9E1B') +
    facet_wrap(~Birth.Year)
```

```r
ggplot(aes(x = 'user.type'), data = subset(chi, !is.na(Birth.Year))) +
    geom_bar(binwidth = 70, color = '#63666A', fill = '#FF9E1B') +
    facet_wrap(~Birth.Year)
```

Looking at the data set for New York, we can see that people who were born between 1990 and 1993 were the highest subscribers for bike share. What is interesting is the bike share subscriptions have gradually dropped starting for people who were born between 1994 to 2001. The reason for this could new technologies that came to market which better served the transportation needs of people born between 1994 to 2001.

Looking at the data set for Chicago, we can see that people who were born between 1990 and 1993 were the highest subscribers for bike share. What is interesting is the bike share subscriptions have gradually dropped starting for people who were born between 1994 to 2016. The reason for this could new technologies that came to market which better served the transportation needs of people born between 1994 to 2016.

Based on the outcome of New York and Chicago, a similar trend could have happened in Washington.

Overall Bike Share is most popular with people born between 1990 and 1993.
